```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
x, y = make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, weights=[0.9], n_redundant=0, random_state=12)
```

```python
x.shape
```

```python
y.shape
```

```python
## Creating the data frame
df1 = pd.DataFrame(x, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])
df = pd.concat([df1, df2], axis=1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['f1'], df['f2'], c=df['target'])
```

### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
## Transform the dataset
oversample = SMOTE()
x, y = oversample.fit_resample(df[['f1', 'f2']], df['target'])
```

```python
x.shape, y.shape
```

```python
y[y==0]
```

```python
len(y[y==0]), len(y[y==1])
```

```python
over_df1 = pd.DataFrame(x, columns=['f1', 'f2'])
over_df2 = pd.DataFrame(y, columns=['target'])
over_df = pd.concat([over_df1, over_df2], axis=1)
```
