```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a customized line plot
plt.plot(x, y, color='red', linestyle='--')
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a customized line plot
plt.plot(x, y, color='blue', linestyle='--', marker='o')
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a customized line plot
plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=3)
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a customized line plot
plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=3, markersize=9)
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a customized line plot
plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=3)
plt.grid(True)
```
